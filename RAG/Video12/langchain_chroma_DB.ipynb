```python
from langchain_google_genai import ChatGoogleGenerativeAI,GoogleGenerativeAIEmbeddings
from langchain_chroma import Chroma
```

```python
from langchain_core.documents import Document

# 1. Define PSL-themed content and rich metadata
docs = [
    Document(
        page_content="Islamabad United is the only team to have won the PSL title three times (2016, 2018, 2024).",
        metadata={"team": "Islamabad United", "city": "Islamabad", "stat_type": "championships", "source": "psl_history.pdf"}
    ),
    Document(
        page_content="Babar Azam is the highest run-scorer in the history of PSL, known for his consistent cover drives.",
        metadata={"player": "Babar Azam", "role": "Batsman", "stat_type": "records", "source": "player_stats.csv"}
    ),
    Document(
        page_content="The 2024 PSL Final was held at the National Bank Stadium in Karachi, where United beat Multan Sultans.",
        metadata={"year": 2024, "venue": "Karachi", "event": "Final", "source": "match_reports"}
    ),
    Document(
        page_content="Shaheen Shah Afridi led Lahore Qalandars to back-to-back titles in 2022 and 2023.",
        metadata={"team": "Lahore Qalandars", "captain": "Shaheen Afridi", "stat_type": "leadership", "source": "psl_archives"}
    ),
    Document(
        page_content="Peshawar Zalmi's 'Yellow Storm' fan base is considered one of the most passionate in the league.",
        metadata={"team": "Peshawar Zalmi", "city": "Peshawar", "category": "fan_culture", "source": "social_media_analysis"}
    )
]

# Verify the Islamabad doc
print(f"Team: {docs[0].metadata['team']}")
print(f"Fact: {docs[0].page_content}")
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model="gemini-embedding-001",google_api_key='')
```

```python
vector_store=Chroma(embedding_function=embeddings,persist_directory='my_chromaDB',collection_name='sample')
```

```python
vector_store.add_documents(docs)
```

```python
vector_store.get(include=['embeddings','documents','metadatas'])
```

```python
vector_store.similarity_search(query="who will psl most time",k=2)
```

```python
vector_store.similarity_search_with_score(query="who will psl most time",k=2)
```
